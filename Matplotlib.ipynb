# MATPLOTLIB -USED TO CREATE PLOTS

```python
import matplotlib.pyplot as plt # to import matplotlib
```

```python
import numpy as np # to get data for plots
```

```python
x=np.linspace(0,10,100)#will give 100 values between 0 to 10 equally spaced
y=np.sin(x)#sin value of all values in x
z=np.cos(x)#cos value of all values of x
```

```python
print(x)
```

```python
print(y)
```

```python
print(z)
```

```python
#lets plot the data 
```

```python
#sin wave
plt.plot(x,y)
plt.show()
```

```python
#cosine wave
plt.plot(x,z)
plt.show()
```

```python
# adding title x-axis &y-axis
plt.plot(x,y)
plt.xlabel('angle')
plt.ylabel('sine value')
plt.title('sin wave')
plt.show()
```

```python
#parabola
x=np.linspace(-10,10,20)
y=x**2
plt.plot(x,y)
plt.show()
```

```python
#here the plot is printed via line but we can also plot using dotted line and different colour
plt.plot(x,y,'r.')#red color and dot
plt.show()
```

```python
plt.plot(x,y,'y*')# yellow color and *dotted
plt.show()
```

```python
plt.plot(x, y, color='green', marker='o', linestyle='--')
```

```python
x=np.linspace(-5,5,50)
plt.plot(x,np.sin(x),'g-')
plt.plot(x,np.cos(x),'r--')
plt.show()
```

# bar plot

```python
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])# enclose our figure in rectanle here 0,0 tells the origin 1,1 is height and width of rectangle
languages=['English','French','Spanish','Latin','German']
people=[100,50,150,40,70]
ax.bar(languages,people,color='green')
plt.xlabel('LANGUAGES')
plt.ylabel('NO OF PEOPLE')
plt.show()
```

# pie chart

```python
fig1=plt.figure()
ax=fig1.add_axes([0,0,1,1])
languages=['English','French','Spanish','Latin','German']
people=[100,50,150,40,70]
ax.pie(people,labels=languages,autopct='%1.1f%%') # will prnt 1float point no
plt.show()
```

# scatterplot

```python
x=np.linspace(0,10,30)
y=np.sin(x)
z=np.cos(x)
fig2=plt.figure()
ax=fig2.add_axes([0,0,1,1])
ax.scatter(x,y,color='g')
ax.scatter(x,z,color='r')
plt.show()
           
```

```python
#3d scatterplot
```

```python
fig3=plt.figure()
ax=plt.axes(projection='3d')
z=20*np.random.random(100)
x=np.sin(z)
y=np.cos(z)
ax.scatter(x,y,z,c=z,cmap='Blues')# c=z means use the z value to determine clor from lighter to darkest
plt.show()
```
